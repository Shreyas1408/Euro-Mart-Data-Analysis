```python
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df
```

<p><span style="color: #999999;">Changing the data format:</span></p>

```python
changing = {'TRUE':'True','FALSE':'False'}
df['Feedback'] = df['Feedback?'].map(changing)
df['Feedback'].astype(bool)
df.drop(columns = ['Feedback?'], inplace = True)
df
```

```python
df.dtypes.to_frame().reset_index()
```

<p><span style="color: #999999;">In this process we have changed the data which were in upper letter format</span></p>

<p><span style="color: #999999;">So we have changed the data which are TRUE and FALSE into bool form,&nbsp;</span><span style="color: #999999;">by using <strong><span style="color: #0000ff;">map</span></strong> key function</span></p>

```python
Discount_in_perc = df['Discount']*100
df['Discount In Percentage'] = Discount_in_perc.astype('int')
df
```

```python
df = df.drop(['Discount'], axis = 1)
df
```

<p>&nbsp; &nbsp; &nbsp; Data visualization is the representation of data through use of common graphics, such as charts, plots, infographics and even animations. These visual displays of information communicate complex data relationships and data-driven insights in a way that is easy to understand.</p>
<p><span style="color: #999999;">Univariate analysis:</span></p>
<ul>
<li><span style="color: #999999;">Univariate analysis&nbsp;explores each variable in a data set, separately.</span></li>
</ul>
<p><span style="color: #999999;">Bivariate analysis:</span></p>
<ul>
<li><span style="color: #999999;">Bivariate analysis is the analysis of two variables to determine relationships between them.</span></li>
</ul>
<p><span style="color: #999999;">Multivariate analysis:</span></p>
<ul>
<li><span style="color: #999999;">Multivariate analysis&nbsp;involves evaluating multiple variables (more than two) to identify any possible association among them.&nbsp;</span></li>
</ul>

```python
high_perf = df[~df['Feedback'].isin(['False'])]
top_feedback_country = high_perf.groupby(['Country'])['Feedback'].value_counts().sort_values(ascending = False).to_frame()

plt.rc('figure',figsize = (8,4))
top_feedback_country.plot( kind = 'barh', color = 'green', legend = False)
plt.xlabel('Feedback with True counts')
plt.show()

display(top_feedback_country.style.background_gradient(cmap="Greens"))
```

```python
least_perf = df[df['Feedback'].isin(['False'])]
least_feedback_country = least_perf.groupby(['Country'])['Feedback'].value_counts().sort_values(ascending = False).to_frame()

plt.rc('figure',figsize = (8,4))
least_feedback_country.plot( kind = 'barh', color = 'red', legend = False)
plt.xlabel('Feedback with False counts')
plt.show()

display(least_feedback_country.style.background_gradient(cmap="Reds"))
```

<p><span style="color: #999999;">In the above cell we have displayed the&nbsp;false /&nbsp;negative feedback by the customers</span></p>

```python
total_feedback = df.groupby(['Country'])['Feedback'].count().sort_values(ascending = False)
improvement_percentage = (least_feedback_country['count']/total_feedback)*100
plt.rc('figure',figsize = (12,4))
improvement_percentage.sort_values().plot(kind = 'barh', 
                                          stacked = False, 
                                          color = ['#FF0000','#FF0050','#FF0100','#FF0250','#FF0500','#FF0750','#FF1000','#FF2000','#FF3000','#FF4000','#FF5000','#FF6000','#FF7000','#FF8000','#FF9000','#F10000']
                                         )
plt.title('Needed Improvement Towards False Feedback')
plt.xlabel('Needed Improvement In Percentage')
plt.show()

improvement_percentage.round().sort_values(ascending = False).to_frame()
```

<p><span style="color: #999999;">In the above cell we have displayed the improvement needed to get 100% positive feedback&nbsp;</span></p>
